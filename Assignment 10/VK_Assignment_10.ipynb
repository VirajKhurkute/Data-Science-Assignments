```python
data = []
for i in range(1, 7501):
    data.append([str(df.values[i, j]) for j in range(0, 20)])
```

```python
print(type(data))
```

**Apriori Algorithm**

```python
asso = apriori(data, min_support=0.0045, min_confidence=0.2, min_lift=3, min_length=2)
results = list(asso)
```

```python
print("There are {} Relation derived.".format(len(results)))
```

**Association Rules**

```python
for i in range(0, len(results)):
    print(results[i][0])
```

**Rules Generated**

```python
for item in results:
    pair = item[0]    # first index of the inner list
    items = [x for x in pair]
    print("Rule: " + items[0] + " -> " + items[1])    # Contains base item and add item
    print("Support: " + str(item[1]))     # second index of the inner list
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("*****************************")
```

```python
sample = [['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Milk', 'Apple', 'Kidney Beans', 'Eggs'],
           ['Milk', 'Unicorn', 'Corn', 'Kidney Beans', 'Yogurt'],
           ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice cream', 'Eggs']]
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from sklearn.feature_extraction.text import TfidfTransformer
from mlxtend.frequent_patterns import apriori
```

**Model Training**

```python
te = TransactionEncoder()
te_try = te.fit(sample).transform(sample)
```

```python
df = pd.DataFrame(te_try, columns=te.columns_)
df
```

```python
apriori(df,min_support=0.5)
```

**Model Training with Column Result return**

```python
apriori(df,min_support=0.5, use_colnames=True)
```

**Calculate the itemsets's length**

```python
itemsets = apriori(df, min_support=0.6, use_colnames=True)
itemsets['length'] = itemsets['itemsets'].apply(lambda x: len(x))
itemsets
```

```python
itemsets[ (itemsets['length'] == 2) & (itemsets['support'] >= 0.8) ]
```

```python
itemsets[itemsets['itemsets'] == {'Onion', 'Eggs'} ]
```

Following model building, it was found that when someone buys kidney beans, he also buys eggs and onions, as we know that association rules define what goes with what.
